```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 데이터 불러오기
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names) # 데이터프레임, columns
labels = pd.DataFrame(iris.target) # 데이터프레임
```

```python
data.info()
```

```python
labels.info()
```

```python
data = pd.concat([data, labels], axis=1)
data.head()
```

#### dbscan

```python
model = DBSCAN(eps=0.5, min_samples=5)
predict = pd.DataFrame(model.fit_predict(data))
predict.columns=['predict']
```

```python
predict
```

```python
r = pd.concat([data, predict], axis=1)
```

```python
r
```

```python
sns.pairplot(r, hue='predict')
plt.show() 
```

```python
sns.pairplot(data, hue=0)
plt.show() #군집분석 별로..좋지 않음
```

### K-Means

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3, random_state=21)
km.fit(data)
```

```python
pred = pd.DataFrame(km.labels_)
pred.columns=['predict']
```

```python
r2 = pd.concat([data, pred], axis=1)
```

```python
r2.info()
```

```python
sns.pairplot(r2, hue='predict')
plt.show() #iris 데이터는 KMeans가 DBSCAN보다 군집 분석 잘 된다
```

- 아이리스 데이터에서 결과가 K-Means가 더 좋아보이지만 모든 데이터에서 K-Means가 군집분석을 잘한다는 의미는 아님
- 머신러닝에서는 무조건적으로 좋다고 말할 수 있는 모델은 없다 -> 각각 다른 데이터들에 대해서 그 데이터를 분석하고 어떤 알고리즘이 사용하기에 적합한지 판단해야함
- 데이터 특성에 맞는 모델을 사용해야함
