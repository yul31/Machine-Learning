```python
import nltk
from nltk.corpus import stopwords
nltk.download('book', quiet=True)
from nltk.book import *
from nltk.tokenize import RegexpTokenizer, WordPunctTokenizer
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# 목록 확인
nltk.corpus.gutenberg.fileids()
```

```python
# 목록 중 가장 첫번째 파일 사용
data = nltk.corpus.gutenberg.raw('austen-emma.txt')
data = data[:3000] #3000개만 사용
```

```python
len(data)
```

```python
# 단어 단위 토크나이저
tokenizer = WordPunctTokenizer() #띄어쓰기 단위 토크나이저
data = tokenizer.tokenize(data)
```

```python
data
```

```python
data = nltk.pos_tag(data) #단어의 품사까지 같이 보고싶을 때 사용하는 토크나이저
```

```python
data
```

```python
# 불용어 제거
data = [each_word for each_word in data if each_word not in stopwords.words()]
```

```python
data
```

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
# 단어별 언급 빈도 확인
data_cnt = Counter(data)
#data_cnt1 = data_cnt.most_common(100) # 가장 많이 나온 100개 단어 -> 코드 오류남

data_cnt = dict(data_cnt) #사전 형태(key, value값)으로
```

```python
data_cnt
```

```python
# 집계된 단어 그래프 표시
data_graph = nltk.Text(data, name='word count')
data_graph.plot() # nltk의 기본적인 그래프 그려주는 기능
```

```python
# 워드클라우드 그리기  #포스태깅(pos_tag) 주석처리 후 실행
wordcloud = WordCloud().generate_from_frequencies(data_cnt)

plt.figure(figsize=(8,4))
plt.imshow(wordcloud)
plt.show()
```
