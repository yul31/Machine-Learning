- 예측값이 0.5보다 작을 경우 0, 0.5보다 클 경우 1로 예측(이진 분류)

#### 와인 데이터를 이용한 이진분류
- 포루투갈의 비뉴 베르드 지방에서 만들어진 와인을 측정한 데이터
- 데이터 셋 구성
    - 레드와인: 1599개/ 화이트와인: 4898개/ 총 6497개
    - 12개의 특성과 1개의 클래스(레이블)로 구성됨

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wine = pd.read_csv('wine.csv')
wine
```

```python
del wine['Unnamed: 0'] #Unnamed 컬럼 삭제
```

```python
wine
```

```python
x = wine.iloc[:,1:]
y = wine['type']
```

```python
x
```

```python
y
```

```python
wine.info()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(C=1, max_iter=1000)
```

```python
lr.fit(x_train, y_train)
```

```python
train_score = lr.score(x_train, y_train)
test_score = lr.score(x_test, y_test)
```

```python
print('train score:', train_score)
print('test score:', test_score)
print('train score - test score :', train_score - test_score)
```

### 최적의 C값 찾기

```python
c_list = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
train_acc = []
test_acc = []
```

```python
for i in c_list :
    lr = LogisticRegression(C=i, max_iter=10000)
    lr.fit(x_train, y_train)
    
    train_score = lr.score(x_train, y_train)
    test_score = lr.score(x_test, y_test)
    
    train_acc.append(train_score)
    test_acc.append(test_score)
```

```python
train_acc
```

```python
test_acc
```

```python
from matplotlib import rc
rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(10,5))
plt.plot(train_acc, label='train')
plt.plot(test_acc, label='test')
plt.title('로지스틱 회귀 모델 C-acc 그래프')
plt.ylabel('acc')
plt.xticks(np.arange(7), c_list)
plt.legend()
plt.grid()
plt.show()
```

### GridSearch

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.001, 0.01, 0.1, 1, 10, 100, 1000]}
```

```python
grid_search = GridSearchCV(LogisticRegression(), param_grid, cv=5)
```

```python
grid_search.fit(x_train, y_train)
```

```python
print('최적 매개변수 :', grid_search.best_params_)
print('최고 교차 검증 점수 :', grid_search.best_score_)
```
